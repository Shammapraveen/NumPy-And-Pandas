## Line plot

```python
import matplotlib.pyplot as plt 
x = [15,82,20,60,90] 
 
y = [25,80,60,32,0] 

plt.plot(x,y) 

plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [1,2,3] 
y = [2,4,6] 

plt.plot(x, y, label = "line1") 

x1 = [1,2,3] 
y1 = [4,9,5] 

plt.plot(x1, y1, label = "line2") 

plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.title('Two lines on one graph') 

plt.legend() 

plt.show()
```

```python
import matplotlib.pyplot as plt 

x = [2,4,6,8,10,12] 

y = [6,9,1,5,2,6] 


plt.plot(x, y, color='red', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='green', markersize=12) 

plt.ylim(1,8) 
plt.xlim(1,8) 

plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.show()
```

## Bar Plot

```python
import matplotlib.pyplot as plt 
x = [15,80,38,6,95] 
 
y = [27,49,69,30,0] 

plt.bar(x,y) 

plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [5,2,3,6,9] 
 
y = [27,48,69,32,100] 

plt.bar(x,y) 

plt.show()
```

```python
import matplotlib.pyplot as plt 

x = [1, 2, 3, 4, 5] 
 
h = [10, 24, 36, 40, 5] 

tick_label = ['one', 'two', 'three', 'four', 'five'] 

plt.bar(x, h, tick_label = tick_label, width = 0.6, color = ['red', 'black']) 
 
plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.title('Bar graph!') 

plt.show()
```

## Histrogram

```python
import matplotlib.pyplot as plt 
y = [20,48,50,80,30] 

plt.hist(y) 

plt.show()
```

```python
import matplotlib.pyplot as plt 
y = [5,2,3,6,10] 

plt.hist(y) 

plt.show()
```

```python
import matplotlib.pyplot as plt 
ages = [2,5,70,40,30,45,50,45,43,40,44,60,7,13,57,18,90,77,32,21,20,40] 

range = (0, 100) 
bins = 10

plt.hist(ages, bins, range, color = 'green',histtype = 'bar', rwidth = 0.8) 

plt.xlabel('Age group') 

plt.ylabel('No. of people affected by corona') 

plt.show()
```

## Pie-chart

```python

import matplotlib.pyplot as plt 
company = ['DevIncept', 'Adobe', 'Infosys', 'Amazon'] 

s = [8, 3, 4, 6] 

colors = ['y', 'g', 'r', 'b'] 

plt.pie(s, labels = company, colors=colors, startangle=90, shadow = True, explode = (0.2, 0, 0, 0), radius = 1.2, autopct = '%1.1f%%') 

plt.legend() 

plt.show()
```
