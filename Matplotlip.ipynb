### Matplotlib
**First we need to import the matplotlib library.**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 4, 8, 16], [1, 10, 2, 13])  # Plot some data on the axes.
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

**An empty figure with no Axes**

```python
fig = plt.figure()
```

 **Afigure with a single Axes**

```python
fig, ax = plt.subplots()
```

**A figure with a 4x4 grid of Axes**

```python
fig, axs = plt.subplots(4, 4)
```

linspace : y = linspace(a,b,n) generates a row vector y of n points linearly spaced between and including a and b.

We can plot many graphs in a single graph:

```python
x = np.linspace(3, 7, 500)

fig, ax = plt.subplots()  
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic') 
ax.plot(x, x**3, label='cubic') 
ax.set_xlabel('x label')  #Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Plot")  # Add a title to the axes.
ax.legend()
```
